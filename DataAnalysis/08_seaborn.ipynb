```python
# 스타일 설정

sns.set_style("ticks") # 스타일 테마 설정
sns.set_context("notebook") # 문맥에 따라 스타일 크기 조정
sns.set_palette("pastel")

plt.rcParams['font.family'] = 'Malgun Gothic' # 맑은 고딕이 설치되어 있는 경우
plt.rcParams['axes.unicode_minus'] = False # 마이너스(-)부호 깨짐 방지
```

```python
# 범주형 데이터 그래프
# countplot
palette = sns.color_palette("pastel")
tips = sns.load_dataset("tips")
sns.countplot(data = tips, x = "day", hue = "sex", palette = palette)
plt.title("요일별 방문자 수")
plt.grid(linestyle = ":")
plt.show()
```

```python
# barplot

palette = sns.color_palette("colorblind")
sns.barplot(data = tips, x = "day", y = "tip", palette = palette)
plt.title("요일별 평균 팁 금액")
plt.show()
```

```python
# displot()
# 히스토그램 + 커널 밀도(kde)
# kde = True  --> kde 그래프
sns.displot(data = tips, x = "total_bill", bins = 20, kde = True)
plt.title("Total bill")
plt.show()
```

```python
# scatterplot()
# 산점도

sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex")
plt.show()
```

```python
# heatmap()
# 상관계수 행렬의 그래프
penguins = sns.load_dataset("penguins")

# 상관계수 행렬
# - 상관계수: 두 변수 사이의 직선 관계가 얼마나 강한지를 숫자로 표현한 값
# - heatmap에서 색으로 표시되는 숫자가 상관계수
# - 상관계수 = 1: 양의 상관관계(하나가 증가하면 다른 것도 같은 비율로 증가)
# ex) 키, 몸무게
# - 상관계수 = -1: 음의 상관관계(하나가 증가하면 다른 변수는 감소)
# ex) 공부시간, 실수

corr = penguins.corr(numeric_only = True)
sns.heatmap(corr, annot = True, fmt = ".2f", cmap = "inferno")
plt.show()
```

```python
# jointplot()
sns.jointplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g", kind = "reg")
plt.show()
```
