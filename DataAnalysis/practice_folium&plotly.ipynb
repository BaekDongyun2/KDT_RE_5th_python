## 1. Folium과 Plotly 이해하기

### Folium은 지도를 만드는 라이브러리
- 최종 결과물: HTML 파일
- 지도, 마커, 팝업 등을 HTML로 만들어줌

### Plotly는 인터랙티브 차트를 만드는 라이브러리
- 최종 결과물: HTML/JavaScript 코드
- 줌, 호버 등 상호작용이 가능한 차트를 만듦

## 2. 왜 HTML로 변환해야 하는가?

- 두 라이브러리가 만나는 지점
[Python 코드] -> [HTML 결과물] -> [웹 브라우저에서 실행]
- Point: Folium도 Ploty도 결국 웹 페이지를 만드는 도구다.

- Folium = 큰 액자(지도)
- Plotly 차트 = 작은 그림

Q: 작은 그림을 큰 액자 안에 어떻게 넣을까?
A: 둘 다 "같은 형식"으로 만들어야 한다. ==> HTML

## 3. 코드로 단계별 확인하기

```python
import plotly.graph_objects as go
import plotly.offline as pyo

# STEP 1: Python에서 Plotly 차트 객체 생성
fig = go.Figure(data=[go.Bar(x=['A', 'B', 'C'], y=[10, 20, 25])])

# 이 시점에서 fig는 Python의 객체
print(type(fig)) # <class 'plotly.graph_objs._figure.Figure'>

# STEP 2: HTML로 변환
chart_html = pyo.plot(fig, include_plotlyjs='cdn', output_type='div')

# 이제 chart_html은 문자열(HTML 코드)임
print(type(chart_html)) # <class 'str'>
print(chart_html[:100]) # (코드의 일부) <div>                        <script type="text/javascript">window.PlotlyConfig = {MathJaxConfig: 'l
```

큰 신문 = Folium 지도

신문 안의 작은 광고 박스 = IFrame

광고 안의 내용 = Plotly 차트

```python
# HTML 파일 생성
import folium
import plotly.graph_objects as go

# 1. Plotly 차트 생성
fig = go.Figure(data = [
    go.Bar(x=['A','B','C'], y=[10, 20, 15])
])
fig.update_layout(
    title='샘플 차트',
    width=400,
    height=300
)

# 2. HTML로 변환
chart_html = fig.to_html(include_plotlyjs='cdn', div_id='myChart')

# 3. Folium 지도 생성
m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# 4. IFrame과 Popup 추가
iframe = folium.IFrame(chart_html, width=450, height=350)
popup = folium.Popup(iframe, max_width=450)

# 5. 마커에 팝업 추가
folium.Marker(
    location=[37.5665, 126.9780],
    popup=popup,
    tooltip='클릭하면 차트가 나타납니다.'
).add_to(m)

m.save('map_with_plotly.html')

m
```

## 엑셀 데이터 → Plotly → Folium 완벽 가이드

[엑셀 파일]

    ↓ pandas로 읽기

[DataFrame (표 형태의 데이터)]

    ↓ Plotly로 시각화

[HTML 문자열]

    ↓ Folium에 삽입
    
[지도 위의 차트]

```python
# 3-1. 기본 변환

# Plotly 차트 -> HTML 문자열
chart_html = fig.to_html(
    include_plotlyjs='cdn', # Plotly 라이브러리를 CDN에서 로드
    div_id='my_chart' # HTML div의 고유 ID
)

# plotly 차트는 javascript 라이브러리가 필요함
# 3가지 옵션
'''
옵션 / 의미 / 장점 / 단점
'cdn' / 인터넷에서 로드 / 파일 크기 작음 / 인터넷 필요
True / HTML에 포함 / 오프라인 가능 / 파일 3MB+ 증가
False / 포함 안함 / 최소 크기 / 차트 안보임
'''

print("HTML 길이: ", len(chart_html), "글자")
print(chart_html[:200])

# 3-2. HTML이 뭔지 이해하기

# 간단한 예제
simple_html = """
<div id="container">
    <h1>제목</h1>
    <p>내용</p>
</div>
"""

print(simple_html)

'''
HTML = 웹 브라우저가 이해하는 언어
<div>: 박스(컨테이너)
<h1>: 제목
Plotly 차트도 결국 <div> 안에 javascript 코드로 변화됨
'''
```

```python
# 4-2. 각 지역에 차트 추가하기

import folium
import plotly.graph_objects as go
import pandas as pd

# 데이터 읽기
df = pd.read_excel('cities_Data.xlsx')

# 지도 생성
m = folium.Map(location=[36.5, 127.5], zoom_start=7)

# 각 지역별 차트 생성 및 추가
for idx, row in df.iterrows():
    city_name = row['지역']
    population = row['인구']
    lat = row['위도']
    lon = row['경도']

    # 해당 지역만의 차트 생성
    fig = go.Figure(data=[
        go.Bar(
            x=['인구'],
            y=[population],
            text=[f"{population}만명"],
            textposition='auto'
        )
    ])

    fig.update_layout(
        title=f"{city_name} 인구",
        width=400,
        height=300,
        showlegend=False
    )

    # HTML로 변환
    chart_html = fig.to_html(include_plotlyjs='cdn')

    # IFrame과 Popup 생성
    iframe = folium.IFrame(chart_html, width=450, height=350)
    popup = folium.Popup(iframe, max_width=450)

    # 마커 추가
    folium.Marker(
        location=[lat,lon],
        popup=popup,
        tooltip=f"{city_name} (클릭하여 차트 보기)",
        icon=folium.Icon(color='blue', icon='info-sign')
    ).add_to(m)

# 저장 및 표시
m.save('korea_cities_chart.html')

m
```

```python
# 도전 과제 1: 평점 막대 차트

import folium
import plotly.graph_objects as go
import pandas as pd

df = pd.read_excel('seoul_cafes.xlsx')

# 지도 생성(서울)
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

for idx, row in df.iterrows():
    # 데이터 추출
    name = row['장소']
    lat = row['위도']
    lon = row['경도']
    rating = row['평점']
    price = row['가격']

    fig = go.Figure(data=[
        go.Bar(
            x=['평점'],
            y=[rating],
            text=[f"{rating}점"],
            textposition='auto',
            marker_color='lightblue'
        )
    ])

    fig.update_layout(
        title=f"{name} 평점",
        width=400,
        height=300,
        yaxis_range=[0, 5],
        showlegend=False
    )

    # HTML 변환 및 지도에 추가
    chart_html = fig.to_html(include_plotlyjs='cdn')
    iframe = folium.IFrame(chart_html, width=450, height=350)
    popup = folium.Popup(iframe, max_width=450)

    # 평점에 따라 마커 색상 변경
    if rating >= 4.5:
        color = "green"
    elif rating >= 3.5:
        color = "blue"
    else:
        color = "orange"

    folium.Marker(
        location=[lat, lon],
        popup=popup,
        tooltip=f"{name} ({rating} 점)",
        icon=folium.Icon(color=color, icon='coffee', prefix='fa')
    ).add_to(m)

    m.save('seoul_cafes_rating.html')

    m
```

```python
# 도전 과제 2: 평점/가격 비교 차트

import folium
import plotly.graph_objects as go
import pandas as pd

df = pd.read_excel('seoul_cafes.xlsx')

# 지도 생성(서울)
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

for idx, row in df.iterrows():
    # 데이터 추출
    name = row['장소']
    lat = row['위도']
    lon = row['경도']
    rating = row['평점']
    price = row['가격']

    # 평점과 가격 함께 비교
    fig = go.Figure()

    # 평점 (왼쪽 축)
    fig.add_trace(go.Bar(
        x=['평점'],
        y=[rating],
        name='평점',
        marker_color='lightblue',
        yaxis='y1',
        text=[f"{rating}점"],
        textposition='auto'
    ))

    # 가격 (오른쪽 축)
    fig.add_trace(go.Bar(
        x=['가격'],
        y=[price],
        name='가격',
        marker_color='lightcoral',
        yaxis='y2',
        text=[f"{price:,}원"],
        textposition='auto'
    ))

    fig.update_layout(
        title=f"{name}",
        width=500,
        height=350,
        yaxis1=dict(title='평점', range=[0, 5]),
        yaxis2=dict(title='가격 (원)', overlaying='y', side='right', range=[0, 8000]),
        showlegend=True
    )

    chart_html = fig.to_html(include_plotlyjs='cdn')
    iframe = folium.IFrame(chart_html, width=550, height=400)
    popup = folium.Popup(iframe, max_width=550)

    folium.Marker(
        location=[lat, lon],
        popup=popup,
        tooltip=f"{name}",
        icon=folium.Icon(color='purple', icpm='info-sign')
    ).add_to(m)

    m.save('seoul_cafes_comparison.html')

    m
```

```python
# 도전 과제 3: 다중 지표 레이더 차트
import folium
import plotly.graph_objects as go
import pandas as pd

df = pd.read_excel('seoul_cafes.xlsx')
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

for idx, row in df.iterrows():
    name = row['장소']
    lat = row['위도']
    lon = row['경도']
    
    # 정규화 (0-5 스케일)
    rating_score = row['평점']
    price_score = 5 - (row['가격'] / 2000) # 가격이 낮을수록 높은 점수
    visitor_score = row['방문자수'] / 100
    seat_score = row['좌석수'] / 30
    hour_score = row['영업시간'] / 5

    # 레이더 차트
    fig = go.Figure()

    fig.add_trace(go.Scatterpolar(
        r=[rating_score, price_score, visitor_score, seat_score, hour_score],
        theta=['평점', '가성비', '인기도', '좌석', '영업시간'],
        fill='toself',
        name=name
    ))

    fig.update_layout(
        polar=dict(
            radialaxis=dict(visible=True, range=[0, 5])
        ),
        title=f"{name} 종합 분석",
        width=450,
        height=400
    )

    chart_html = fig.to_html(include_plotlyjs='cdn')
    iframe = folium.IFrame(chart_html, width=500, height=450)
    popup = folium.Popup(iframe, max_width=500)

    folium.Marker(
        location=[lat,lon],
        popup=popup,
        tooltip=f"{name}",
        icon=folium.Icon(color='red', icon='chart-area', prefix='fa')
    ).add_to(m)

    m.save('seoul_cages_radar.html')

    m
```

```python
# 도전 과제 4: 히트맵 스타일 (방문자 수 색상)
import folium
from folium.plugins import HeatMap
import pandas as pd

df = pd.read_excel('seoul_cafes.xlsx')

# 히트맵용 지도
m = folium.Map(location=[37.5665, 126.9780], zoom_start=11)

# 히트맵 데이터 준비
heat_data = [[row['위도'], row['경도'], row['방문자수']] for idx, row in df.iterrows()]

# 히트맵 추가
HeatMap(heat_data, radius=20, blur=25, max_zoom=13, name='방문자 밀집도').add_to(m)

for idx, row in df.iterrows():
    name = row['장소']
    lat = row['위도']
    lon = row['경도']
    visitors = row['방문자수']
    rating = row['평점']

    # 간단한 차트 생성
    fig = go.Figure(data=[
        go.Bar(
            x=['방문자수', '평점'],
            y=[visitors, rating*100], # 평점을 100배 해서 스케일 맞춤
            text=[f"{visitors}명", f"{rating}점"],
            textposition='auto' 
        )
    ])

    fig.update_layout(
    title=f"{name}",
    width=400,
    height=300
    )
    
    chart_html = fig.to_html(include_plotlyjs='cdn')
    iframe = folium.IFrame(chart_html, width=450, height=350)
    popup = folium.Popup(iframe, max_width=450)

    # 마커 크기를 방문자 수에 비례하게 표시
    folium.CircleMarker(
        location=[lat, lon],
        radius=visitors/50,
        popup=popup,
        tooltip=f"{name} 방문자: {visitors}명 {rating}점",
        color='darkred',
        fill=True,
        fillColor='red',
        fillOpacity=0.7
    ).add_to(m)

# 레이어 컨트롤 추가 (히트맵 on/off 가능)
folium.LayerControl().add_to(m)

m.save('seoul_cafes_heatmap.html')

m
```
