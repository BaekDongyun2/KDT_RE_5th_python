## 09. Folium
+ 지도 시각화 라이브러리

```python
import folium
```

```python
# 기본 지도 생성
m = folium.Map(location = [37.51355584833682, 127.10420263921695], zoom_start = 18)

# 기본 마커
folium.Marker([37.51355584833682, 127.10420263921695], popup = "Subway", tooltip = "롯데월드 몰", icon = folium.Icon(color = "black", icon = "fa-solid fa-bus", prefix = "fa-solid")).add_to(m)

# 원형 마커
folium.CircleMarker([37.51355584833682, 127.10420263921695], radius = 100, color = "#adcdff", fill_color = "#1c73ff", popup = "CircleMarker", tooltip = "tooltip").add_to(m)

# 클릭한 곳에 마커 추가
m.add_child(folium.ClickForMarker(popup = "내가 클릭한 곳"))

# 클릭한 곳의 위도와 경도
m.add_child(folium.LatLngPopup())

m
```

```python
# 실습 1.강의장 주변 맛집, 카페, 공원 등등 원하는 곳 3군데 이상 표시하기

# 기본 지도 생성
MAP = folium.Map(location = [37.6092965, 126.91577238783975], zoom_start = 18)

# 기본 마커
folium.Marker([37.6092965, 126.91577238783975],
              popup = "Classroom", 
              tooltip = "강의실", 
              icon = folium.Icon(color = "black", icon = "fa-solid fa-book-open-reader", 
              prefix = "fa-solid")).add_to(MAP)

folium.Marker([37.6126, 126.9176],
               popup = "Restaurant",
               tooltip = "식당",
               icon = folium.Icon(color = "black", icon = "fa-solid fa-bowl-rice",
               prefix = "fa-solid")).add_to(MAP)

folium.Marker([37.6119, 126.9175],
               popup = "Gym",
               tooltip = "헬스장",
               icon = folium.Icon(color = "black", icon = "fa-solid fa-dumbbell",
               prefix = "fa-solid")).add_to(MAP)

folium.Marker([37.6196, 126.9211],
               popup = "Cafe",
               tooltip = "카페",
               icon = folium.Icon(color = "black", icon = "fa-solid fa-mug-hot",
               prefix = "fa-solid")).add_to(MAP)               

MAP

```

```python
# 실습 2. 택시 승차대 위치 표시하기
import pandas as pd
df = pd.read_csv("./dataset/택시승차대 현황.csv", encoding = 'cp949')

MAP = folium.Map(location = [37.6092965, 126.91577238783975], zoom_start = 18)

for i in range(len(df)):
    folium.Marker(
        location = [df.loc[i, "위도"], df.loc[i, "경도"]],
                popup = f"{df.loc[i, '위치명']} ({df.loc[i, '도로명 주소']})",
                tooltip = df.loc[i, "지번 주소"],
                icon = folium.Icon(color = "black", icon = "fa-solid fa-taxi",
                prefix = "fa-solid")).add_to(MAP)
MAP
```
