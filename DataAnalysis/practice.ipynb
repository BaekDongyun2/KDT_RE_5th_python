```python
# 1. 넘파이 모듈을 np라는 이름으로 임포트해보세요.
import numpy as np
```

```python
# 2. 리스트를 넘파이의 ndarray 객체로 변환해보세요.
data = [1, 2, 3]
dt = np.ndarray(data)
print(dt)
```

```python
# 3. 0부터 99까지의 정수가 저장된 ndarray 객체를 만들어 보세요.
np.ndarray(100)
```

```python
# 4. 구구단 3단의 결과가 저장된 ndarray 객체를 만들어 보세요
arr = np.arange(1,10) * 3
print(arr)
```

```python
# 5. 1000보다 작은 짝수가 저장된 ndarray 객체를 만들어 보세요.
arr = np.arange(1000)
arr[arr % 2 == 0]
# np.arange(2, 1000, 2)
```

```python
# 6. 0부터 1까지 0.1 간격으로 구성된 ndarray를 생성해보세요. 1은 포함되지 않습니다.
arr = np.arange(0, 1, 0.1)
print(arr)
```

```python
# 7. 리스트에는 6개의 데이터가 저장돼 있습니다. 2행 3열의 ndarray로 변환하세요.
data = [0, 1, 2, 3, 4, 5]
arr = np.array(data)
arr.reshape(2, 3)
```

```python
# 8. 다음의 2차원 ndarray 1차원 ndarray로 변환하세요.
arr = np.array([
    [0, 1, 2],
    [3, 4, 5]
])
arr.flatten()
```

```python
# 9. 주어진 ndarray를 3행 1열로 변환하세요.
arr = np.arange(3)
arr.reshape(3, 1)
```

```python
# 10. 에러의 원인을 설명하세요.
# > np.arange(6).reshape(2, 4)
# ValueError: cannot reshape array of size 6 into shape (2,4)

# --> arange(6)에 의해 총 6개의 데이터가 저장된 ndarray가 생성된다. 이를 2행 4열의 행렬로 변환하려 했기 때문에 데이터의 갯수가 달라 에러가 발생한 것이다.
```
