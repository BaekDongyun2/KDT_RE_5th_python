### Series
- 인데스(라벨)가 붙은 1차원 배열
- 인덱스 이름을 직접 지정할 수 있음

```python
# series 생성
# 1. 리스트로 생성
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
```

```python
# 2. 인덱스를 직접 지정
s = pd.Series(data, index = ["a", "b", "c", "d"])
print(s)
```

```python
# 3. 딕셔너리로 생성
data = {'서울':100, '부산':200, '대구':300}
s = pd.Series(data)
print(s)
```

```python
# 4. 하나의 값으로 series 생성
s = pd.Series(7, index = ["가", "나", "다"])
print(s)
```

```python
# 시리즈의 주요 속성
data = [10, 20, 30, 40]
s = pd.Series(data, index = ["a", "b", "c", "d"], name = 'test data')
print("values: ", s.values) # 실제 데이터 배열
print("index: ", s.index) # 인덱스 객체
print("dtype: ", s.dtype) # 실제 데이터의 타입
print("size: ", s.size) # 데이터의 개수
print("shape: ", s.shape) # 배열의 형태
print("name: ", s.name) # series 이름
```

```python
# 시리즈의 연산
s1 = pd. Series([10, 20, 30], index = ["a", "b", "c"])
s2 = pd.Series([5, 6, 7], index = ['c', 'b', 'd'])

result = s1 + s2

print(result)
```

```python
# 브로드캐스팅
s = pd.Series([5, 10, 15])

print(s + 3)
print(s * 2)
```

```python
# 불리언 연산 및 인덱싱
s = pd.Series([50, 60, 70, 80, 90, 100])

mask = s > 70
print(mask)
print(s[mask])
```

```python
# 실습 1-1
s = pd.Series([5, 10, 15, 20])
print(s)
```

```python
# 실습 1-2
s = pd. Series([90, 80, 85, 70], index = ["국어", "영어", "수학", "과학"])
print(s)
```

```python
# 실습 1-3  
data = {'서울':950, '부산':340, '인천':520}
s = pd.Series(data)
print(s["인천"])
```

```python
# 실습 1-4
s = pd.Series([1, 2 ,3, 4])
print("데이터 타입: ", s.dtype)
```

```python
# 실습 1-5
s1 = pd.Series([3, 5, 7], index=['a', 'b', 'c'])
s2 = pd.Series([10, 20, 30], index=['b', 'c', 'd'])

result = s1 + s2

print(result)
```

```python
# 실습 1-6
s = pd.Series([1, 2, 3, 4, 5])
print(s + 10)
```

### DataFrame
- 2차원 라벨 데이터 구조
- 시리즈를 모아놓은 것
- 행과 열이 있고, 행과 열에 대해 라벨을 부여할 수 있음

```python
# DataFrame 생성
# 1. 딕셔너리로 생성
data = {
    "이름": ['kim', 'lee', 'sim'],
    "나이": [15, 23, 44],
    "도시": ["서울", "부산", "대구"]

}

df = pd.DataFrame(data)
df
```

```python
# 2. 리스트의 리스트(2차원 리스트)로 생성
data = [[1, "a"], [2, "b"], [3, "c"]]
df = pd.DataFrame(data, columns = ["번호", "코드"])
df
```

```python
# 3. 딕셔너리의 리스트 생성
data = [
    {"이름": "동윤2", "나이": 26, "도시": "포항"},
    {"이름": "배칠수", "나이": 43, "도시": "서울"}
]

df = pd.DataFrame(data, index = ["a", "b"])
df
```

### 인덱싱과 슬라이싱

```python
# 인덱싱: 컬럼에 대해 인덱싱 적용
# 특정 컬럼(시리즈) 선택
df["이름"]

# 여러 컬럼을 리스트로 선택
df[['이름', '나이']]

# .으로도 조회가 가능함
df.이름
```

### iloc, loc

```python
# iloc
# Integer Location
# 정수 위치 기반의 인덱싱, 슬라이싱
# numpy의 슬라이싱과 거의 유사함

# 단일 행/열 선택
df.iloc[0] # 첫 번째 행
df.iloc[:, 1] # 두 번째 열
```

```python
# 여러 행/열 동시 선택
df.iloc[1:4]
df.iloc[:, 0:2]
df.iloc[1:4, 0:2]
```

```python
# Fancy Indexing

df.iloc[[0,2,4]]
df.iloc[:, [1, 2]]
df.iloc[[0, 2, 4], [1, 2]]
```

```python
# 음수 인덱스 슬라이싱
df.iloc[-1]
df.iloc[:, -2:]
```

```python
# loc
# Location의 약자
# 라벨의 이름 기준으로 인덱싱/슬라이싱
# 단일 행/열 선택
# 시작과 끝을 모두 포함
# 음수 인덱스 사용 X

df.loc[0]
df.loc[:, "이름"]
```

```python
# 여러 행/열 선택
df.loc[2:4, ["이름", "직업"]]
df.loc[2:4, "이름": "직업"]
```

### 통계함수(집계함수)

```python
data = {
    "상품명": ["무선 이어폰", "스마트 워치", "텀블러", "노트북", "블루투스 스피커", "무드등"],
    "가격": [129000, 250000, 15000, 1200000, 85000, 22000],
    "재고": [23, 12, 54, 5, 17, 31]
}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
# 평균
df["가격"].mean()

# 중강값
df["재고"].median()

# 표준편차
df["가격"].std()
df["재고"].std()

# 분산
df["재고"].var()

# 값의 개수
df["상품명"].count()

# 최댓값
df["가격"].max()

# 최솟값
df["재고"].min()

# 합계
df["가격"].sum()
```

```python
# 쵀댓값의 위치와 최소값의 위치
df.idxmax()

df.idxmin()
df["가격"].idxmin()
```

### 결측값 처리
- 결측값: 값이 기록되지 않은 상태의 데이터
- NaN(Not a Number)

```python
data = {
    "이름": ["서준", "하은", "민준", "서연", "이안", "지민"],
    "나이": [22, 28, np.nan, 31, 27, 24],
    "점수": [89, np.nan, 83, 90, 88, 93]
}
df = pd.DataFrame(data)
df
```

```python
# 결측값 탐지
# isnull: 결측값이 맞으면 True 아니면 False
df.isnull()

# 각 컬럼의 결측값의 수 계산
df.isnull().sum()

# 데이터의 전체 결측값 수 계산
df.isnull().sum().sum()
```

```python
# notnull 결측값이 아니면 True, 맞으면 False
df.notnull()
```

```python
# dropna 결측값이 있는 행 삭제
df2 = df.dropna()
df2
```

```python
# 결측값이 있는 열 삭제
df3 = df.dropna(axis = 1)
df3           
```

```python
# fillna 결측값을 특정 값으로 대체
df4 = df.fillna(0)
df4
```

```python
avg_age = df["나이"].mean()
avg_age

df["나이"] = df["나이"].fillna(avg_age)
df
```

```python
# 이전 값으로 결측값 채우기 (forward fill)
df5 = df.ffill()
df5
```

```python
# 뒤의 값으로 결측값 채우기 (backward fill)
df6 = df.bfill()
df6    
```
